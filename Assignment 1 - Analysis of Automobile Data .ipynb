```python
#1
df.replace('?', np.nan, inplace = True)
numeric_columns = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[numeric_columns] = df[numeric_columns].astype(float)

for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)

df.head()
```

```python
#Advanced Statistical Analysis
```

```python
numerical_attributes_cols=df.iloc[:,[0,1,9,10,11,12,13,16,18,19,20,21,22,23,24,25]]
numerical_attributes_cols.head()
```

```python
#1 - multivariate mean vector
mean_vector=np.mean(numerical_attributes_cols, axis=0)
print(mean_vector)
```

```python
#2 - Sample Covariance Matrix using the Inner Product
center=numerical_attributes_cols-numerical_attributes_cols.mean()
cov_matrix=1/numerical_attributes_cols.shape[0]*np.dot(center.T, center)
print(cov_matrix)
```

```python
#3 - Sample Covariance Matrix using the Outer Product
cov_matrix_outer=np.zeros((center.shape[1], center.shape[1]))
for row in center:
    cov_matrix_outer=cov_matrix_outer+np.outer(row,row)

cov_matrix_outer=cov_matrix_outer/center.shape[0]
print(cov_matrix_outer)
```

```python
#4 - Correlation between wheel-base and length
wheel_base=df['wheel-base']
length=df['length']

wheel_base_centered = wheel_base - np.mean(wheel_base)
length_centered = length - np.mean(length)

dot_product = np.dot(wheel_base_centered, length_centered)
magnitude_wheel_base = np.linalg.norm(wheel_base_centered)
magnitude_length = np.linalg.norm(length_centered)

correlation = dot_product / (magnitude_wheel_base * magnitude_length)

print(correlation)
```

```python
#4 - Scatterplot of wheel-base and length
plt.scatter(x=wheel_base_centered,y=length_centered)
plt.xlabel("Wheel Base")
plt.ylabel("Length")
plt.title("Relation between wheel base and length")
```

```python
#5 - Covariance matrix of fuel-type, body-style, and drive-wheels
categorical_columns=['fuel-type','body-style','drive-wheels']

encoded_data = np.array([LabelEncoder().fit_transform(df[col]) for col in categorical_columns]).T

cov_mat = np.cov(encoded_data, rowvar=False)
print(cov_mat)
```

```python
#6 - Probability density function of engine-size
engine_size = df['engine-size']

mean = np.mean(engine_size)
std_dev = np.std(engine_size)

x_values = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)

pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)

plt.plot(x_values, pdf_values, label='PDF', color='blue')
plt.title("Probability Density Function of 'engine-size'")
plt.xlabel("Engine Size")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#7 - Variance Analysis
variances = numerical_attributes_cols.var()

max_variance_attr = variances.idxmax()
min_variance_attr = variances.idxmin()

print("Attribute with the largest variance:", max_variance_attr, "(Variance:", variances[max_variance_attr],")")
print("Attribute with the sallest variance:", min_variance_attr, "(Variance:", variances[min_variance_attr],")")
```

```python
#8 - Covariance Analysis

covariance = numerical_attributes_cols.cov()

cov_values = covariance.where(np.triu(np.ones(covariance.shape), k=1).astype(bool))
max_cov = cov_values.stack().idxmax()
min_cov = cov_values.stack().idxmin()

print("Largest covariance:", max_cov, "Value:",covariance.loc[max_cov])
print("Smallest covariance:", min_cov, "Value:",covariance.loc[min_cov])
```
